```python
from cityscapes import CityScapes
import matplotlib.pyplot as plt
import numpy as np
import torchvision.transforms as T
import torchvision.transforms.functional as F
root_dir = "F:\COMP90055\GMIDA\datas\CityScapes"
dataloader = CityScapes(root_dir,batch_size=1,split='train')
resize = T.Resize((256,512),interpolation=F.InterpolationMode.NEAREST) 
for img_id,image,label in dataloader:
    print(img_id)
    plt.imshow((image[0]).permute(1,2,0).numpy().astype(np.float32))
    plt.show()
    plt.imshow(label[0].numpy())
    plt.show()
    print(np.unique(label[0]))
    small_label = resize(label)
    plt.imshow(small_label[0].numpy())
    plt.show()
    print(np.unique(small_label[0]))
    
    break
```
